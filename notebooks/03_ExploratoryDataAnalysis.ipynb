```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
processed_data = pd.read_csv("../data/03_Processed/02_ProcessedData.csv")
```

```python
# Making stacked bar plot of target distribution

processed_data["target"].value_counts(normalize=True).to_frame().T.plot.barh(stacked=True)

# save as png

plt.savefig("../figures/03_TargetDistribution.png")
```
